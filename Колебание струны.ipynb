```python
#!/usr/bin/env python
# -*- coding: utf-8
import numpy as np
import matplotlib.pyplot as plt # подключаемые библиотеки
import math
  
  
h=0.1                        #шаг по x  
t=0.05                       #шаг по t 
a=1                          #отношение начального натяжения T к массе единице длинны
l=(a*t/h)**2                 #расчётный коэффициент
n=20                         #Количество в сетке узлов по x (i = 0, 1, ... n) 
m=20                         #Колиество слоёв по t (j = 0, 1, ... m).
distanceBeetwenLayers = 6    #промежуток можду слоями 
x=[h*i for i in range(0,n)]  # массив x-ов

U= np.zeros([m,n])           # матрица для размещения расчётных значений
for j in range(0,m):         #внесение условий         
         for i in range(0,n):                  
                  U[0,i]= x[i]-50       #начальные условия                                   
                  U[1,i]=U[0,i]         #начальные условия    
                  U[j,0]=0              #граничные условия   
                  U[j,n-1]=0            #граничные условия   
                    
for j in range(1,m-1):                  #внесение решения (метод сеток)     
       for i in range(1,n-1):
                U[j+1,i]=2*(1-l)*U[j,i]+l*(U[j,i+1]+U[j,i-1])-U[j-1,i]


plt.title(f'Решение задачи на {m} слоях,\n с промежутком {distanceBeetwenLayers}', size=12)  

for i in range(0,m, distanceBeetwenLayers):         #отрисовка слоёв
         plt.plot(x, U[i],label=f"слой {i}")        
        
plt.legend()
plt.grid()
plt.show()   
```
